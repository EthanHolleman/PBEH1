Drop rows that are not unique reads

```python
# chat GPT
def remove_duplicate_rows(df, key_col):
    """
    Identifies rows in a pandas dataframe that have a duplicate key value,
    and returns a new dataframe with only one row for each key value.

    Parameters:
    df (pandas.DataFrame): The dataframe to search for duplicates.
    key_col (str): The name of the column containing the key value.

    Returns:
    pandas.DataFrame: A dataframe containing only one row for each key value.
    """
    new_df = df.drop_duplicates(subset=key_col, keep='first')
    return new_df
```

```python
test_df = remove_duplicate_rows(test_df, 'uniqueID')
```
